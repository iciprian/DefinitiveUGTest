```python
#IMPORTS
import pandas as pd
import numpy as np
import plotly.express as px
pd.set_option('display.max_rows', 1000)
```

```python
#DATOS
path = "/Users/icipriano/Repositorios/Data/"
```

```python
folder = path + "RTWFeb20/"
pname = folder + "stats.csv"
df_All = pd.read_csv(pname)
df_All = df_All[df_All["FINISHED_OK"]==True]
```

```python
def formatear(entrada):
    if entrada == entrada and type(entrada) == str :
        return sum(map(float, entrada.replace('[', '').replace(']', '').split(", ")))
    else :
        return entrada

def recover_val_a(x):
    return [int(s) for s in x.split('_') if s.isdigit()][0]

def recover_val_b(x):
    return [int(s) for s in x.split('_') if s.isdigit()][1]

def recover_val_c(x):
    return [int(s) for s in x.split('_') if s.isdigit()][2]

def recover_val_d(x):
    return float([s for s in x.split('_')][-1])

def recover_val_e(x):
    return [s for s in x.split('_')][0]

def recover_name(x):
    word = [s for s in x.split('_')][0]
    return word.split('.')[0]
```

```python
df_all = df_All.copy()
df_all.reset_index(drop=True, inplace=True)

df_all['NAME']=df_all['INPUT_NAME'].transform(recover_name)
df_all['A']=df_all['PARAMS_FILE'].transform(recover_val_a)
df_all['B']=df_all['PARAMS_FILE'].transform(recover_val_b)
df_all['C']=df_all['PARAMS_FILE'].transform(recover_val_c)
df_all['D']=df_all['PARAMS_FILE'].transform(recover_val_d)
df_all['E']=df_all['PARAMS_FILE'].transform(recover_val_e)

df_all['BZ_TIME'] = [formatear(df_all['BZ_SOLVETIME'][i]) for i in range(len(df_all))]
df_all['LS_TIME'] = [formatear(df_all['TOPOSORT_TIME'][i]) for i in range(len(df_all))]

df_all['pBZ_TIME'] =df_all['BZ_TIME']/df_all['TOTAL_TIME']*100
df_all['pTopo_TIME'] =df_all['LS_TIME']/df_all['TOTAL_TIME']*100
df_all = df_all[['NAME','A','B','C','D','FINAL_OBJVAL','TOTAL_TIME','pBZ_TIME','pTopo_TIME']]

df_all = df_all.round(2)
df_all.index.name = 'run'
df_all.sort_values(by = 'FINAL_OBJVAL', ascending = False)
```
