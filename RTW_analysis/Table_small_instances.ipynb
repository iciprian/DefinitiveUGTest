```python
def formatear(entrada):
    if entrada == entrada and type(entrada) == str :
        return sum(map(float, entrada.replace('[', '').replace(']', '').split(", ")))
    else :
        return entrada

def recover_val_a(x):
    return [int(s) for s in x.split('_') if s.isdigit()][0]

def recover_val_b(x):
    return [int(s) for s in x.split('_') if s.isdigit()][1]

def recover_val_c(x):
    return [int(s) for s in x.split('_') if s.isdigit()][2]

def recover_val_d(x):
    return float([s for s in x.split('_')][-1])

def recover_val_e(x):
    return [s for s in x.split('_')][0]

def recover_name(x):
    word = [s for s in x.split('_')][0]
    return word.split('.')[0]
```

```python
df_Catan = df_Catan.copy()

df_Catan['NAME']=df_Catan['INPUT_NAME'].transform(recover_name)
df_Catan['A']=df_Catan['PARAMS_FILE'].transform(recover_val_a)
df_Catan['B']=df_Catan['PARAMS_FILE'].transform(recover_val_b)
df_Catan['C']=df_Catan['PARAMS_FILE'].transform(recover_val_c)
df_Catan['D']=df_Catan['PARAMS_FILE'].transform(recover_val_d)
df_Catan['E']=df_Catan['PARAMS_FILE'].transform(recover_val_e)

df_Catan['BZ_TIME'] = [formatear(df_Catan['BZ_SOLVETIME'][i]) for i in range(len(df_Catan))]
df_Catan['LS_TIME'] = [formatear(df_Catan['TOPOSORT_TIME'][i]) for i in range(len(df_Catan))]

df_Catan['pBZ_TIME'] = 100*df_Catan['BZ_TIME']/df_Catan['TOTAL_TIME']
df_Catan['pLS_TIME'] = 100*df_Catan['LS_TIME']/df_Catan['TOTAL_TIME']
df_Catan['pother_TIME'] = 100 - df_Catan['pLS_TIME']-df_Catan['pBZ_TIME']

df_Catan['TOTAL_TIME'] = df_Catan['TOTAL_TIME']#/3600

df_Catan['FINAL_OBJVAL'] = df_Catan['FINAL_OBJVAL']#/959082.799

df_Catan = df_Catan[['NAME','A','B','C','D','E','FINAL_OBJVAL','TOTAL_TIME','pBZ_TIME','pLS_TIME','pother_TIME']]
df_Catan = df_Catan.round(2)
df_Catan.index.name = 'run'
df_Catan.sort_values(by = 'FINAL_OBJVAL', ascending = False)
```
